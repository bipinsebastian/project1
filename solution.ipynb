```python
 # Add title and labels
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import csv
from datetime import datetime

def plot_temperature(file_path, linewidth=2, show_low_temp=False):
    # Initialize lists to store the dates and temperatures
    dates = []
    high_temperatures = []
    low_temperatures = []

    # Read the CSV file
    with open(r'C:\Users\Gymnasiet\floeder for python\data-hw-bipinsebastian\projects\project 1\datasets\sitka_weather_07-2021_simple.csv', 'r') as csv_file:
        reader = csv.reader(csv_file)
        header_row = next(reader)  # Skip the header row
        
        # Identify the correct indices for date and temperatures
        date_index = header_row.index('DATE')
        high_temp_index = header_row.index('TMAX')
        low_temp_index = header_row.index('TMIN')
        
        # Read each row and extract dates and temperatures
        for row in reader:
            try:
                current_date = datetime.strptime(row[date_index], '%Y-%m-%d')
                high_temp = int(row[high_temp_index])
                low_temp = int(row[low_temp_index])
            except ValueError:
                print(f"Missing data for {row[date_index]}")
                continue
            dates.append(current_date)
            high_temperatures.append(high_temp)
            low_temperatures.append(low_temp)
    
    # Plot the high temperatures
    plt.figure(figsize=(10, 6))
    plt.plot(dates, high_temperatures, color='red', linewidth=linewidth, label='High Temp')
    
    # Optionally plot the low temperatures
    if show_low_temp:
        plt.plot(dates, low_temperatures, color='blue', linewidth=linewidth, label='Low Temp')

    # Add title and labels
    plt.title(f"Temperature Data - {file_path}", fontsize=16)
    plt.xlabel("Dates", fontsize=12)
    plt.ylabel("Temperatures (F)", fontsize=12)
    
    # Display legend if both temperatures are shown
    if show_low_temp:
        plt.legend()
    
    # Rotate the x-axis labels for better readability
    plt.xticks(rotation=45)
    
    # Adjust layout to avoid overlapping elements
    plt.tight_layout()
    
    # Show the plot
    plt.show()

# Example usage of the function
plot_temperature('sitka_weather_07-2021_simple.csv', linewidth=3, show_low_temp=True)

```
